```python
import heapq

def a_star(graph, start_node, goal_node, heuristic):

    priority_queue = [(0, start_node)]
    cost_so_far = {start_node: 0}
    parent = {start_node: None}

    while priority_queue:

        _, current_node = heapq.heappop(priority_queue)

        if current_node == goal_node:
            path = []
            while current_node:
                path.append(current_node)
                current_node = parent[current_node]

            return path[::-1]

        for neighbor, edge_cost in graph[current_node]:

            new_cost = cost_so_far[current_node] + edge_cost

            if neighbor not in cost_so_far or new_cost < cost_so_far[neighbor]:

                cost_so_far[neighbor] = new_cost

                estimated_cost = new_cost + heuristic[neighbor]

                heapq.heappush(priority_queue, (estimated_cost, neighbor))

                parent[neighbor] = current_node

    return None


graph = {
    'S': [('A', 3), ('B', 2)],
    'A': [('C', 4), ('D', 1)],
    'B': [('E', 3), ('F', 1)],
    'E': [('H', 5)],
    'F': [('I', 2), ('G', 3)],
    'C': [], 'D': [], 'H': [], 'I': [], 'G': []
}

heuristic = {
    'S': 13, 'A': 12, 'B': 4, 'C': 7, 'D': 3,
    'E': 8, 'F': 2, 'H': 4, 'I': 9, 'G': 0
}

start_node = input("Enter Start Node: ")
goal_node = input("Enter Goal Node: ")

path = a_star(graph, start_node, goal_node, heuristic)

print("Path:", path)
```

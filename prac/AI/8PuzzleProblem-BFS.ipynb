```python
from collections import deque
```

```python
N=3
```

```python
goal=[
    [0,1,2],
    [3,4,5],
    [6,7,8]
]
```

```python
row = [0,0,-1,1]
col = [-1,1,0,0]
```

```python
def isGoal(board):
    return board == goal
```

```python
def isValid(x,y):
    return 0<=x<N and 0<=y<N
```

```python
def printBoard(board):
    for r in board:
        print(*r)
    print("--------")
```

```python
def printSolution(path):
    print("\n Solution Path: \n")

    for steps, board in enumerate(path):
        print(f"Steps {steps}")
        printBoard(board)
    print(f"Total Moves {len(path)-1}")
```

```python
def solvePuzzleBFS(start,x,y):
    queue = deque()
    visited = set()

    queue.append((start,x,y,0,[start]))
    visited.add(tuple(map(tuple,start)))
    nodes_explored = 0
    while queue:
        board , x, y, depth, path = queue.popleft()
        nodes_explored +=1

        if isGoal(board):
            print("\nGoal Found\n")
            print(f"Depth {depth}")
            print(f"nodes visited {nodes_explored}")
            printSolution(path)

            return
        for i in range(4):
            newx = x+row[i]
            newy = y+col[i]

            if isValid(newx,newy):
                newBoard = [r[:] for r in board]
                newBoard[x][y], newBoard[newx][newy] =\
                    newBoard[newx][newy], newBoard[x][y]
                
                board_tuple = tuple(map(tuple,newBoard))
                if board_tuple not in visited:
                    visited.add(board_tuple)
                    new_path = path + [newBoard]
                    queue.append(
                    (newBoard, newx,newy,depth+1,new_path)
                    )
    print("Solution Not Found")
```

```python
startBoard = [
    [1,2,3],
    [4,0,5],
    [6,7,8]
]
x,y=1,1
```

```python
solvePuzzleBFS(startBoard,x,y)
```

```python
rules = [
    (["A", "B"], "C"),
    (["C"], "D"),
    (["D"], "E")
]

facts = {"A", "B"}

def forward():
    while True:
        added = False

        for cond, result in rules:
            if all(x in facts for x in cond) and result not in facts:
                facts.add(result)
                print("Inferred:", result)
                added = True

        if not added:
            break

def backward(goal):
    if goal in facts:
        return True

    for cond, result in rules:
        if result == goal:
            return all(backward(x) for x in cond)

    return False

print("Initial Facts:", facts)

print("\nForward Chaining")
forward()
print("Final Facts:", facts)

print("\nBackward Chaining")
goal = "E"

if backward(goal):
    print(goal, "is PROVED")
else:
    print(goal, "is NOT PROVED")
```

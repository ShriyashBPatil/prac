```python
import math

board = [""] * 9

wins = [
    [0,1,2],[3,4,5],[6,7,8],
    [0,3,6],[1,4,7],[2,5,8],
    [0,4,8],[2,4,6]
]

def print_board():
    for i in range(0, 9, 3):
        print((board[i] or " ") + "|" +
              (board[i+1] or " ") + "|" +
              (board[i+2] or " "))

def winner(player):
    for line in wins:
        if all(board[i] == player for i in line):
            return True
    return False

def draw():
    return "" not in board

def minimax(is_ai):

    if winner("X"):
        return 1
    if winner("O"):
        return -1
    if draw():
        return 0

    if is_ai:
        best = -math.inf

        for i in range(9):
            if board[i] == "":
                board[i] = "X"
                best = max(best, minimax(False))
                board[i] = ""

        return best

    else:
        best = math.inf

        for i in range(9):
            if board[i] == "":
                board[i] = "O"
                best = min(best, minimax(True))
                board[i] = ""

        return best

def best_move():
    best_score = -math.inf
    move = -1

    for i in range(9):
        if board[i] == "":
            board[i] = "X"
            score = minimax(False)
            board[i] = ""

            if score > best_score:
                best_score = score
                move = i

    return move

while True:

    print_board()

    pos = int(input("Enter position (0-8): "))

    if board[pos] != "":
        print("Invalid Move")
        continue

    board[pos] = "O"

    if winner("O"):
        print_board()
        print("You Win!")
        break

    if draw():
        print("Draw!")
        break

    ai = best_move()
    board[ai] = "X"

    print("\nAI Move:")
    print_board()

    if winner("X"):
        print("AI Wins!")
        break

    if draw():
        print("Draw!")
        break
```

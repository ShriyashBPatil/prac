```python
import random

dist = [
    [0, 10, 15, 20],
    [10, 0, 35, 25],
    [15, 35, 0, 30],
    [20, 25, 30, 0]
]

def tour_cost(tour):
    cost = 0
    for i in range(len(tour)):
        cost += dist[tour[i]][tour[(i + 1) % len(tour)]]
    return cost

tour = [0, 1, 2, 3]
random.shuffle(tour)

best_cost = tour_cost(tour)

for _ in range(1000):
    new_tour = tour[:]

    i, j = random.sample(range(4), 2)
    new_tour[i], new_tour[j] = new_tour[j], new_tour[i]

    new_cost = tour_cost(new_tour)

    if new_cost < best_cost:
        tour = new_tour
        best_cost = new_cost

print("Best Tour:", tour)
print("Total Distance:", best_cost)
```
